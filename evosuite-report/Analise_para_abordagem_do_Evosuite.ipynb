### Análise da Meta de Cobertura Atingida:
Análise da Meta de Cobertura Atingida: Avalie quantos dos objetivos de cobertura total foram alcançados em cada execução. Compare a quantidade de metas de cobertura total cobertas em cada execução para entender a consistência da cobertura dos testes.


*   Este código carrega o arquivo CSV, calcula a proporção de metas de cobertura cobertas em cada execução, calcula a média da cobertura para cada TARGET_CLASS e também calcula o desvio padrão da cobertura para cada TARGET_CLASS. Isso pode ajudar a entender a consistência da cobertura dos testes para diferentes classes de destino.


```python
import pandas as pd

# Carregar o arquivo CSV
df = pd.read_csv('statistics.csv')

# Calcular a quantidade de metas de cobertura total cobertas em cada execução
df['Coverage_Ratio'] = df['Covered_Goals'] / df['Total_Goals']

# Agrupar os dados por TARGET_CLASS e calcular a média da cobertura
coverage_by_target_class = df.groupby('TARGET_CLASS')['Coverage_Ratio'].mean()

# Exibir a média de cobertura para cada TARGET_CLASS
print("Média de cobertura por TARGET_CLASS:")
print(coverage_by_target_class)

# Comparar a quantidade de metas de cobertura total cobertas em cada execução
# Para entender a consistência da cobertura dos testes, você pode calcular
# a variação ou desvio padrão da cobertura para cada TARGET_CLASS
coverage_std_by_target_class = df.groupby('TARGET_CLASS')['Coverage_Ratio'].std()

# Exibir o desvio padrão da cobertura para cada TARGET_CLASS
print("\nDesvio padrão da cobertura por TARGET_CLASS:")
print(coverage_std_by_target_class)
```

### Análise da Média do Tamanho dos Arquivos de Teste:
Análise da Média do Tamanho dos Arquivos de Teste: Calcule a média do tamanho dos arquivos de teste (Size) em cada execução e analise como essa média se comporta ao longo das repetições. Isso pode ajudar a entender se há alguma tendência de aumento ou diminuição na complexidade dos testes ao longo do tempo.

```python
import pandas as pd

# Carregar o arquivo CSV
df = pd.read_csv("statistics_30RepDefault.csv")

# Agrupar os dados por TARGET_CLASS
grouped = df.groupby('TARGET_CLASS')

# Criar uma lista para armazenar os dados das médias
mean_sizes_data = []

# Iterar sobre cada grupo
for name, group in grouped:
    # Calcular a média do tamanho dos arquivos de teste para cada TARGET_CLASS
    mean_size = group['Size'].mean()
    # Adicionar os dados da média à lista
    mean_sizes_data.append({'TARGET_CLASS': name, 'Mean_Size': mean_size})

# Criar um DataFrame a partir da lista de dados das médias
mean_sizes = pd.DataFrame(mean_sizes_data)

# Exibir as médias calculadas
print(mean_sizes)

# Agora você pode analisar como as médias de tamanho dos arquivos se comportam ao longo das repetições para cada TARGET_CLASS
```

```python
#!/usr/bin/python

import pandas as pd
from scipy.stats import wilcoxon
from numpy import mean, std # version >= 1.7.1 && <= 1.9.1
from math import sqrt

# From http://stackoverflow.com/questions/21532471/how-to-calculate-cohens-d-in-python
def cohen_d(x,y):
  return (mean(x) - mean(y)) / sqrt((std(x, ddof=1) ** 2 + std(y, ddof=1) ** 2) / 2.0)


df = pd.read_csv('statistics_30RepDefaultBranc.csv')

cat1 = df[df['configuration_id']=='Default']
cat2 = df[df['configuration_id']=='Branch']

for column in ['Size', 'Length', 'MutationScore']:
  print ("%s: %.2f (%f)" % (column, cohen_d(cat1[column], cat2[column]), wilcoxon(cat1[column], cat2[column]).pvalue))
```

### Análise da Meta de Cobertura Atingida:
Análise da Meta de Cobertura Atingida: Avalie quantos dos objetivos de cobertura total foram alcançados em cada execução. Compare a quantidade de metas de cobertura total cobertas em cada execução para entender a consistência da cobertura dos testes.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Passo 1: Carregar os dados do arquivo CSV
df = pd.read_csv('statistics_30RepDefault.csv')

# Passo 2: Calcular a porcentagem de cobertura atingida em relação ao total de metas para cada execução
df['Coverage_Percentage'] = (df['Covered_Goals'] / df['Total_Goals']) * 100

# Verificar e remover valores duplicados (se houver)
df = df.drop_duplicates()

# Passo 3: Plotar um gráfico de barras para comparar a cobertura atingida em cada execução
executions = range(1, len(df) + 1)  # Usando o comprimento do DataFrame atualizado
plt.figure(figsize=(10, 6))
plt.bar(executions, df['Coverage_Percentage'], color='blue')
plt.title('Cobertura de Teste por Execução')
plt.xlabel('Execução')
plt.ylabel('Porcentagem de Cobertura Atingida')
plt.grid(True)
plt.show()
```

### Análise da Distribuição do Tamanho dos Arquivos de Teste:

Análise da Distribuição do Tamanho dos Arquivos de Teste: Examine a distribuição do tamanho dos arquivos de teste (Size) ao longo das 30 execuções para identificar padrões ou anomalias. Isso pode fornecer insights sobre a complexidade dos testes gerados em diferentes iterações.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Passo 1: Carregar os dados do arquivo CSV
df = pd.read_csv('statistics_30RepDefault.csv')

# Passo 2: Extrair os tamanhos dos arquivos de teste
sizes = df['Size']

# Passo 3: Plotar o histograma da distribuição
plt.figure(figsize=(10, 6))
plt.hist(sizes, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribuição do Tamanho dos Arquivos de Teste')
plt.xlabel('Tamanho do Arquivo')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Passo 1: Carregar os dados do arquivo CSV
df = pd.read_csv('statistics_30RepDefault.csv')

# Passo 2: Dividir os dados com base na coluna 'TARGET_CLASS'
classes = df['TARGET_CLASS'].unique()

# Passo 3: Plotar o histograma da distribuição para cada classe
plt.figure(figsize=(12, 8))

for target_class in classes:
    sizes = df[df['TARGET_CLASS'] == target_class]['Size']
    plt.hist(sizes, bins=10, edgecolor='black', alpha=0.7, label=target_class)

plt.title('Distribuição do Tamanho dos Arquivos de Teste por Classe')
plt.xlabel('Tamanho do Arquivo')
plt.ylabel('Frequência')
plt.legend(title='Classe')
plt.grid(True)
plt.show()
```
